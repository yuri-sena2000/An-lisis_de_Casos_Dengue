```python
import pandas as pd
```

```python
query = '''
SELECT region, SUM(casos) AS total_casos
FROM Casos_Dengue
GROUP BY region
'''
```

```python
df = pd.read_sql_query(query, conn)
```

```python
conn.close()
```
